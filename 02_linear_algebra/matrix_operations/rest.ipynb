# Задача 1

```python
def dot(a, b):
    out = 0
    for n, v in enumerate(a):
        out += v*b[n]
    return out

dot((0.7, 1.9, 4, 3.2, 5.1, 2.7, 4.8), 
    (21, 40, 29, 31, 18, 31, 26))
```

# Задача 2

```python
def distance(a, b):
    return pow(dot(a, a) + dot(b, b) - 2*dot(a, b), 0.5)

print(round(distance( (4, 7, 2, 1, 6, 1, 3, 3, 2, 8), 
         (2, 3, 8, 8, 8, 8, 7, 1, 0, 5)), 2))
```

# Задача 3

```python
def distance(a, b):
    out = 0
    for n, v in enumerate(a):
        out += abs(v - b[n])
    return out

distance((4, 7, 2, 1, 6, 1, 3, 3, 2, 8),  (2, 3, 8, 8, 8, 8, 7, 1, 0, 5))
```

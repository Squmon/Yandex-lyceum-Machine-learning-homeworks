```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
help(datasets.make_classification)
```

```python
F, C = datasets.make_classification(
    10000,
    2,
    n_classes=4,
    n_informative=2,
    n_clusters_per_class=1,
    n_redundant=0,
    n_repeated=0)
pl.scatter(F[0:,0], F[0:,1], c = C, s = 2)
pl.show()
```

```python
F_train, F_test, C_train, C_test = train_test_split(F, C)
```

```python
KN = KNeighborsClassifier()
KN.fit(F_train, C_train)

C_predicted = KN.predict(F_test)
```

```python
pl.subplot(2, 2, 1)
pl.scatter(F[0:,0], F[0:,1], c = C, s = 2)
pl.title("Оригинал")

pl.subplot(2, 2, 2)
pl.scatter(F_train[0:,0], F_train[0:,1], c = C_train, s = 2)
pl.title("на чем тренировали")

pl.subplot(2, 2, 3)
pl.scatter(F_test[0:,0], F_test[0:,1], c = C_test, s = 2)
pl.title("на чем проверяли")


pl.subplot(2, 2, 4)
pl.scatter(F_test[0:,0], F_test[0:,1], c = C_predicted, s = 2)
pl.title("что выдало в итоге")

pl.show()
```

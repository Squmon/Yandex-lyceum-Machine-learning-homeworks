## Математические операции

Мат операции очевидно те же что и в обычном, однако для произведения векторов и матриц идут исключения:

```python
A = np.arange(10)
B = np.arange(10)

A*B
```

Формы массивов не обязательно должны совпадать, главное что бы они были broadcast

```python
a = np.array([[1, 2, 3], [2, 3, 4]])
b = np.array([3, 4, 5])
c = np.array([1, 2])
print(a+b)
# print(a + c) но это не работает
```

```python
a = np.eye(10, 10)
b = np.arange(10)
a*b
```

```python
N = 10
a = np.eye(N, N)
b = np.arange(1, N + 1)[::-1]
a*b**2
```

```python
# @ матричное умножение

A = np.array([1, 2, 3])
M = np.eye(3, 3)
M = M[:1,:2]
print(M)
```

```python
a = np.ones([3, 2, 2])
b = np.ones([3, 2, 2])
(a@b).shape
```

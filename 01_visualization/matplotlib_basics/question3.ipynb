Имеется 1000 фотографий с автомобилями, припаркованными на улицах города. 750 автомобилей припаркованы согласно правилам дорожного движения и разметке, а 250 - нарушают. Классификатор учится предсказывать факт нарушения (1 - нарушил правила, 0 - все хорошо). Классификатор дал правильный ответ на 700 правильно припаркованных автомобилях и на 150 нарушителях. Посчитайте F-меру данного классификатора (\beta=1). Ответ запишите в виде десятичной дроби, при необходимости округлите ответ до двух знаков после разделителя, в качестве разделителя используйте запятую.

```python
N = 1000
positive_count = 750 # количество правильных
negative_count = 250 # количество не правильных

TP = 700
TN = 150

FP = positive_count - TP
FN = negative_count - TN
```

![Alt text](image-2.png)

```python
def Prec(TP, FP):
    return TP/(TP + FP)

def Recall(TP, FN):
    return TP/(TP + FN)

def Fscore(Precision, Recall):
    return 2 * (Precision*Recall)/(Precision+Recall)
```

```python
round(Fscore(Prec(TP, FP), Recall(TP, FN)), 2)
```
